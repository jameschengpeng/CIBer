```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file = "Glass change_train_size_acc.xlsx"
df = pd.read_excel(file, header=None)
```

```python
ciber_mean = list(df.iloc[2:9,13].values)
ciber_mean = [1-x for x in ciber_mean]
ciber_up = list(df.iloc[2:9,12].values)
ciber_up = [1-x for x in ciber_up]
ciber_low = list(df.iloc[2:9,11].values)
ciber_low = [1-x for x in ciber_low]

NB_mean = list(df.iloc[11:18,13].values)
NB_mean = [1-x for x in NB_mean]
NB_up = list(df.iloc[11:18,12].values)
NB_up = [1-x for x in NB_up]
NB_low = list(df.iloc[11:18,11].values)
NB_low = [1-x for x in NB_low]

xgb_mean = list(df.iloc[20:27,13].values)
xgb_mean = [1-x for x in xgb_mean]
xgb_up = list(df.iloc[20:27,12].values)
xgb_up = [1-x for x in xgb_up]
xgb_low = list(df.iloc[20:27,11].values)
xgb_low = [1-x for x in xgb_low]

rf_mean = list(df.iloc[29:36,13].values)
rf_mean = [1-x for x in rf_mean]
rf_up = list(df.iloc[29:36,12].values)
rf_up = [1-x for x in rf_up]
rf_low = list(df.iloc[29:36,11].values)
rf_low = [1-x for x in rf_low]

dt_mean = list(df.iloc[38:45,13].values)
dt_mean = [1-x for x in dt_mean]
dt_up = list(df.iloc[38:45,12].values)
dt_up = [1-x for x in dt_up]
dt_low = list(df.iloc[38:45,11].values)
dt_low = [1-x for x in dt_low]

svm_mean = list(df.iloc[47:54,13].values)
svm_mean = [1-x for x in svm_mean]
svm_up = list(df.iloc[47:54,12].values)
svm_up = [1-x for x in svm_up]
svm_low = list(df.iloc[47:54,11].values)
svm_low = [1-x for x in svm_low]

lr_mean = list(df.iloc[56:63,13].values)
lr_mean = [1-x for x in lr_mean]
lr_up = list(df.iloc[56:63,12].values)
lr_up = [1-x for x in lr_up]
lr_low = list(df.iloc[56:63,11].values)
lr_low = [1-x for x in lr_low]
```

```python
def plot_mean_and_CI(name, x, mean, low, up, color_mean = None, color_shading=None):
    plt.plot(x, mean, color = color_mean, label = name)
    plt.fill_between(x, low, up, color=color_shading, alpha=.5)
```

```python
x = ["10%","20%","30%","40%","50%","60%","70%"]
fig = plt.figure(1, figsize=(7, 5))
plot_mean_and_CI("CIBer", x, ciber_mean, ciber_low, ciber_up, color_mean = '#007FFF', color_shading='#007FFF')
# plot_mean_and_CI("Naive Bayes", x, NB_mean, NB_low, NB_up, color_mean = 'goldenrod', color_shading='goldenrod')
# plot_mean_and_CI("XGBoost", x, xgb_mean, xgb_low, xgb_up, color_mean = 'grey', color_shading='grey')
plot_mean_and_CI("Random Forest", x, rf_mean, rf_low, rf_up, color_mean = 'navy', color_shading='navy')
# plot_mean_and_CI("Decision Tree", x, dt_mean, dt_low, dt_up, color_mean = 'goldenrod', color_shading='goldenrod')
# plot_mean_and_CI("SVM", x, svm_mean, svm_low, svm_up, color_mean = 'grey', color_shading='grey')
# plot_mean_and_CI("Logistic Regression", x, lr_mean, lr_low, lr_up, color_mean = 'navy', color_shading='navy')
plt.xlabel("Training Percentage")
plt.ylabel("Out-of-Sample Error Rate")
plt.title("Error Rate at Different Training Percentage")
lgd = plt.legend(bbox_to_anchor=(1.05, 1.0, 0.3, 0.2), loc='upper left')
fig.savefig('err_change_glass_rf.png', bbox_extra_artists=(lgd,), bbox_inches='tight', dpi = 800)
```
