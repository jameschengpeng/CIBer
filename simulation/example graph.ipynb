```python
import matplotlib.pyplot as plt
```

```python
class0_start = [0,20]
class0_end = [100,120]
class1_start = [20,0]
class1_end = [120,100]
```

```python
class0_x = [class0_start[0], class0_end[0]]
class0_y = [class0_start[1], class0_end[1]]
class1_x = [class1_start[0], class1_end[0]]
class1_y = [class1_start[1], class1_end[1]]
```

```python
plt.plot(class0_x, class0_y, label='Class 0')
plt.plot(class1_x, class1_y, label='Class 1')
plt.legend()
plt.xlabel('X0')
plt.ylabel('X1')
plt.savefig('example.png', dpi = 800)
```
